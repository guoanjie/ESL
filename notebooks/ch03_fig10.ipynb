```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LassoLars
from datasets.prostate import X_train, y_train
```

```python
reg = LassoLars(alpha=0)
reg.fit(X_train, y_train)
xx = np.sum(np.abs(reg.coef_path_), axis=0)
xx /= xx[-1]
```

```python
ax = pd.DataFrame(
    data=reg.coef_path_.T, index=xx, columns=X_train.columns,
).plot(style='.-')
ax.axvline(x=xx[3], color='red', linestyle='--', linewidth=1)
ax.set_xlabel("Shrinkage Factor s")
ax.set_ylabel("Coefficients");
```
