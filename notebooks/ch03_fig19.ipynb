```python
import numpy as np
import pandas as pd

from datasets.prostate import X_train, y_train
from models import ForwardStagewiseRegression
```

```python
reg = ForwardStagewiseRegression(eps=.01)
reg.fit(X_train, y_train)
ax = pd.DataFrame(
    data=reg.coef_path_,
    columns=X_train.columns,
).plot()
ax.set_xlabel('Iteration')
ax.set_ylabel('Coefficients')
ax.set_title('FS$_\epsilon$');
```

```python
reg = ForwardStagewiseRegression(eps=1e-3)
reg.fit(X_train, y_train)
xx = np.sum(np.abs(reg.coef_path_), axis=1)
ax = pd.DataFrame(
    data=reg.coef_path_,
    index=xx,
    columns=X_train.columns,
).plot()
ax.set_xlabel('$L_1$ Arc-length of Coefficients')
ax.set_ylabel('Coefficients')
ax.set_title('FS$_0$');
```
